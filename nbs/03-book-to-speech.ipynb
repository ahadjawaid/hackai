```python
lines = get_gutenberg_book(book_id)
lines
```

```python
text = " ".join(lines)
```

```python
def sublists_within_char_limit(words, char_limit):
    current_sublist = L([])
    current_length = 0

    for word in words:
        if current_length + len(word) + len(current_sublist) <= char_limit:  # + len(current_sublist) for spaces between words
            current_sublist.append(word)
            current_length += len(word)
        else:
            yield current_sublist
            current_sublist = L([word])
            current_length = len(word)

    if current_sublist:
        yield current_sublist
```

```python
char_limit = 1000
words = text.split(" ")
wordIter = sublists_within_char_limit(words, char_limit)
```

```python
def wav_generator(wordIter, num_chars, char_limit):
    total_char_len = len(" ".join(words))
    for _ in range(total_char_len//char_limit):
        text = " ".join(next(wordIter))
        wav, _ = synthesize_speech(text)
        yield wav
```
